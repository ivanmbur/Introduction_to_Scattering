The problem one faces here is that, for a given radius $R^2$, it is difficult to know if there exists an $\vec{n}\in\mathbb{Z}^3$ such that $\vec{n}^2=R^2$. An example for which there is no such solution is $R^2=7$. Instead, in order to find the energy levels, the easiest thing to do is to try to list as uniquely as possible a vector $\vec{n}$ for each energy level. In order to do this one can explit the symmetries of the problem to fix $\vec{n}\in\mathbb{N}^3$ and $n_x\geq n_y\geq n_z$. This still doesn't give a unique vector for each radius, e.g. $(3,0,0)$ and $(2,2,1)$ have the same energy. 

Another approach might be to use the fact that a positive integer is the sum of three squares if and only if it is not of the form $R^2=4^k(8m+7)$, for integers $k,m\geq 0$. It is easy to see that such an $R^2$ cannot be a sum of three squares. To see this, note that every square modulo 4 is either 0 (if it is even $(2k)^2=4k^2=0\mod 4$) or 1 (if it is odd $(2k+1)^2=4k^2+4k+1=1\mod 4$). Similarly, every square module 8 is either 0, 4 (if it is even since it has to be 0 modulo 4) or 1 (if it is odd since $(2k+1)^2=4k(k+1)+1=1\mod 8$ given that $k(k+1)$ is even). Now, if $n_x^2+n_y^2+n_z^2=R^2=8m+7$, reduction modulo 8 would mean that one can sum 7 out of a sum of three integers in $\{1,4\}$. This is clearly impossible. On the other hand, if $n_x^2+n_y^2+n_z^2=4^k(8m+7)$ with $k>0$, reduction modulo 4 shows that each of these squares is a multiple of 4, i.e. all of the integers $n_x$, $n_y$ and $n_z$ are even. Thus, dividing the whole equation by 4, we have $\left(\frac{n_x}{2}\right)^2+\left(\frac{n_y}{2}\right)^2+\left(\frac{n_z}{2}\right)^2=4^{k-1}(8m+7)$, i.e. we have expressed $4^{k-1}(8m+7)$ as the sum of three squares. Continuing in this fashion one can reduce back to the case $k=0$ and conclude the same contradiction as before.

The other implication is much more difficult to proof and I don't know it. So let's not do it.

Finally, let us note that the presentation $4^k(8m+7)$ is unique. Indeed, if $4^k(8m+7)=4^{k'}(8m+7)$ and $k>k'$, we would have that $4^{k-k'}(8m+7)=(8m'+7)$, i.e. an even number equal to an odd number. We must then have $k=k'$, from which we immediately deduce that $m=m'$.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['text.usetex'] = True
plt.rc('text.latex', preamble=r'\usepackage{amsmath}\usepackage{physics}')

# Maximum energy level disallowed

kmax = 0
mmax = 1
Nmax = 4**kmax*(8*mmax + 7) 

# Energy levels

N = np.arange(0, Nmax + 1, 1)
to_delete = [4**k*(8*m + 7) for k in range(kmax + 1) for m in range(mmax + 1)]
N = np.delete(N, to_delete)
    
# Energy as a function of length and level

def E(L, Rsq):
    return np.sqrt(1 + (2*np.pi/L)**2*Rsq)

fig, ax = plt.subplots()
L = np.linspace(0.5, 10, 100)
for Rsq in N:
    ax.plot(L, E(L,Rsq), label = r"$\vec{{n}}^2={}$".format(Rsq))
ax.set_xlabel(r"$mL$")
ax.set_ylabel(r"$E/m$")
ax.set_title(r"Energy levels in the CM frame")
ax.legend()
    
```

Out of the CM frame, the energy level is not fully determined by $\vec{n}^2$ anymore.
